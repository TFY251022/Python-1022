條件語句 if-elif-else

```python
'''
條件判斷
if ...
if ... else ...
if ... elif ... else ...

'''
```

##### if

```python
num=10
if num > 5:
    print("num 大於 5")

```

##### if...else...

```python
name = "banana"
if name == "apple":
    print("name is apple")
else:
    print("name is not apple")
```

##### if...elif... (else)

```python
num =5
if num >10:
    print("num 大於 10")
elif num % 2 == 0:
    print("num 是偶數")
elif type(num) == int:
    print("num 資料型態為 int")
else:
    print("num 沒有大於10 且不為 int")
```
