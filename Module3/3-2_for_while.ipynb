| 用法 | 說明 |
|:---:|:---:|
| `range(m)` | 走訪 0 到 m-1 的數字 |
| `range(n, m)` | 走訪 n 到 m-1 的數字  |
| `range(n, m, step)` | 以step為間距，走訪 n 到 m-1 的數字 |

##### for

```python
n = 1
for char in "contents":
    print(char, end="")
    print(n, end=",")
    n += 1
```

```python
# range(m) 走訪0到m-1的數字
for i in range(5):
    print(i)
```

```python
for i in range(5):
    print("Hello World!")
```

```python
# range(n, m) 走訪n到m-1的數字
for i in range(1,11):
    print(i, end=", ")

```

```python
# range(n, m, step) 以step為間距,走訪n到m-1的數字
for i in range(1,11,2):
    print(i, end=", ")
```

### 小試身手 for  
請產出 a 開頭的 n 個偶數(含 a 本身)  
Ex:  
a = 5, n = 3  
Output : 6 8 10

a = 4, n = 5  
Output : 4 6 8 10 12

```python
for i in range(6,11,2):
    print(i, end=" ")
```
