```python
# f-string (又稱作 formatted string literals, version >= 3.6)
name = "Leo"
age = 18

print(f"{name} is {age} years old.")
```

```python
now = "2025-11-01 15:54"
status = "INFO"
msg = "ABC"
log = f"{now} [{status}] - {msg}"
print(log)
```

```python
# 靠左對齊、靠右對齊
name = "Leo"
age = 18

msg = f"name: [{name:<10}], age: [{age:>10}]"
print(msg)
```

```python
# 靠左對齊、靠右對齊，字數不足，補「ㄏ」
# (沒寫補什麼，預設空格)
name = "Leo"
age = 18

msg = f"name: [{name:X<10}], age: [{age:0>10}]"
print(msg)
```

```python
# 置中對齊
name = "Leo"
age = 18

msg = f"name: [{name:^10}], age: [{age:^6}]"
print(msg)
```

```python
# 指定文字長度上限 (只有文字，才在格式化字串中加入「.」來限定字串長度)
name = "Leo"

msg = f"name: [{name:.2}]"
print(msg)

```

```python
data = 3.14159
print(f"{data:.2f}")
```
