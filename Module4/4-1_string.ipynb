|string方法| 功能 |
|:---:|:---:|
|str.replace() | 替換字串 |
|str.strip() | 去除兩側空格 |
|str.upper() | 字串小寫變大寫 |
|str,lower() | 字串大寫變小寫 |

```python
# 字串變數初始化
name = "jarvis"
print(name)
#索引取得對應字元:使用中括號
print(name[0])
```

```python
# 替換字串
# 方法 也是 function 的一種
# 定義在 class 底下的 function 可以稱為 method
'''
str.replace()
將 string 中的 str1 替換成 str2
'''

name = "Alex"
new_name = name.replace("ex", "ice")
print(new_name)
```

```python
# 去除兩側空格
'''
str.strip()
去除字串 string 左、右兩邊的空格
'''

data = "  |__d a t a__|  "
print(data)
#去除字串兩邊空格
new_data = data.strip()
print(new_data)
#去除文字中間空格
new_data = new_data.replace(" ","")
print(new_data)
```

```python
data = "1  |__d a t a__|  1"
print(data)
#去除字串左邊1
new_data = data.lstrip("1")
print(new_data)
#去除字串右邊1
new_data = data.rstrip("1")
print(new_data)
```

```python
# 字串小寫變大寫
'''
str.lower()
將字串 string 裡的字母全部改成小寫
'''
data = "CAR"
new_data = data.lower()
print(new_data)

'''
str,upper()
將字串 string 裡的字母全部改成大寫
'''
print("good".upper())
```

```python
# 字串分割成 list
#str.split()
greeting = "hello world! hello John"
str_list = greeting.split("!")
print(str_list)
```

```python
greeting = "hello world! hello John"
str_list = greeting.split(" ")
print(str_list)
for list_str in str_list:
    for char in list_str:
        print(char, end=", ")
```
