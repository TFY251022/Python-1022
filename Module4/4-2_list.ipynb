```python
# 初始化一個 list
ids = [1, 2, 3, 4, 5, 6]
print(ids)
```

```python
# 新增元素在 list 尾端
# list.append()
ids = [1, 2, 3, 4, 5, 6]
ids.append(7)
print(ids)

```

```python
# 查詢索引位置的元素
# 由0起算
ids = [1, 2, 3, 4, 5, 6]
print(ids[3])
```

```python
# 修改索引位置的元素
ids = [1, 2, 3, 4, 5, 6]
ids[3] = 99
print(ids)

```

```python
# 刪除索引位置的元素
# list.pop(<index索引位置>)
ids = [1, 2, 3, 4, 5, 6]
ids.pop(3)
print(ids)

```

```python
# 在指定索引 1 的位置新增元素 9  ，其餘元素往後移
# list.insert()
ids = [1, 2, 3, 4, 5, 6]
ids.insert(1,9)
print(ids)
```

```python
# 移除指定的值:一次只刪除一個
# list.remove(<value值>)
ids = [1, 2, 3, 4, 5, 6]
ids.remove(4)
print(ids)


```

```python
#刪除多個同元素
idr = [1, 2, 3, 4, 5, 6]
for i in range(5):
    idr.insert(1,9)
print(idr)  
#元素9在list裡   
while 9 in idr:
        idr.remove(9)
print(idr)
```

```python
# 計算 list 內的某個值 共有幾個
# list.count(<value值>)
ids = [1, 2, 3, 4, 5, 6]
for i in range(5):
    ids.append(10)
print(ids)

cnt_10 = ids.count(10)
print(cnt_10)

```

```python
# list 的合併
# list.extend()
ids1 = [1, 2, 3, 4]
ids2 = [5, 6, 7]
ids1.extend(ids2)
print(ids1)

total_ids = ids1 + ids2
print(total_ids)
```

### list 搭配 for 迴圈

```python
# 建立一個 1~10 的 list
idt = []
for i in range(1,11):
    idt.append(i)
print(idt)
```

```python
# 建立一個 1~50 的 list 只要偶數
idt = []
for i in range(1,51):
    if i % 2 == 0:
        idt.append(i)
print(idt)
```

```python
# 建立一個list 裡面要放遞增的 list
# 例如 [[1], [1, 2], [1, 2, 3], ..., [1, 2, ...,10]]
ids_out =[]
for i in range(1,11):
    ids_in = []
    print(f"i : {i}")
    for c in range(1,i+1):
        print(f"c : {c}")
        ids_in.append(c)
    print(ids_in,id(ids_in))
    ids_out.append(ids_in)
print(ids_out)
```

```python
# 建立一個list 裡面要放遞增的 list
# list.copy()
ids_out = []
ids_in = []
for i in range(1,11):
    ids_in.append(i)
    print(ids_in,id(ids_in.copy()))
    ids_out.append(ids_in.copy())
print(ids_out)
```

```python
# enumerate() 同時取得 index 與 value
# 回傳 tuple元組
ids = [1, 2, 3, 4, 5]
for i in enumerate(ids):
    print(i)

ids_1 = [10, 2, 35, 4, 15]
for index, value in enumerate(ids_1):
    print(f"index: {index}, value: {value}")

```

### List comprehension 列表推導式
`new_list = [expression for item in iterable [if condition]]`  
  
`list的名稱 = [表達式 for 元素 in 迭代對象 [if 判斷式(...的條件)]]`

```python
# 建立一個 1~10 的 list
idt = []
for i in range(1,11):
    idt.append(i)
print(idt)
```

```python
# list comprehension 縮寫
num_list_1_to_10 = [i for i in range(1, 11)]
print(num_list_1_to_10)
```

```python
# 1~10的二次方
data = [i**2 for i in range(1, 11)]
print(data)
```

```python
# 建立一個 1~50 的 list 只要偶數
idt = []
for i in range(1,51):
    if i % 2 == 0:
        idt.append(i)
print(idt)

idt_com = [i for i in range(1, 51) if i % 2 ==0]
print(idt_com)
```

```python
# 建立一個list 裡面要放遞增的 list
# 例如 [[1], [1, 2], [1, 2, 3], ..., [1, 2, ...,10]]
ids_out =[]
for i in range(1, 11):
    ids_in = []
    for c in range(1, i+1):
        ids_in.append(c)
    ids_out.append(ids_in)
print(ids_out)

ids_out = []
for i in range(1,11):
    ids_out.append([c for c in range(1, i+1)])
    #ids_in = [c for c in range(1, i+1)]
    #ids_out.append(ids_in)    
print(ids_out)

ids_out = []
ids_out = [[c for c in range(1, i+1)] for i in range(1, 11)]
print(ids_out)
```
