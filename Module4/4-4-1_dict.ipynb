```python
# 初始化 dict
fruit_price_dict = {
    "apple": 100, 
    "orange": 50, 
    "tomato": 30
    }
print(fruit_price_dict)
```

```python
# 印出 蘋果的價格
fruit_price_dict = {
    "apple": 100, 
    "orange": 50, 
    "tomato": 30
    }
print(fruit_price_dict["apple"])
```

```python
# 修改 橘子的價格
fruit_price_dict = {
        "apple": 100, 
        "orange": 50, 
        "tomato": 30
        }
fruit_price_dict["orange"] = 70
print(fruit_price_dict)
```

```python
# 刪除 番茄
fruit_price_dict = {
        "apple": 100, 
        "orange": 50, 
        "tomato": 30
        }
fruit_price_dict.pop("tomato")
print(fruit_price_dict)

fruit_price_dict = {
    "apple": 100, 
    "orange": 50, 
    "tomato": 30
    }
del(fruit_price_dict["tomato"])
print(fruit_price_dict)
```

```python
# 建立一個 value 含有 dict 的 dict
user_info = {
    "name": "Jarvis",
    "age": 20,
    "info":{
        "nickname": "Lu",
        "favorite": "ironman",
        "phone": [
            "0912345678",
            "0987654321"
        ]
        
    }
}

print(user_info)
```

```python
# 取得 nickname
print(user_info["info"]["nickname"])
```

```python
# 取得所有 phone_number
print(user_info["info"]["phone"])

for phone_number in user_info["info"]["phone"]:
    print(phone_number)

```

```python
'''
改寫 fruit_dict 每個水果需要 price, quantity/amount, color
apple 有 綠色 和 紅色
orange 有 橘色
tomato 有 紅色
'''
# 對應的資料結構建議保持一致屬性
fruit_dict = {
        "apple": {
            "price": 30,
            "amount": 100,
            "color": ["red", "green"]
        }, 
        "orange": {
            "price": 50,
            "amount": 30,
            "color": ["orange"]
        },  
        "tomato": {
            "price": 100,
            "amount": 20,
            "color": ["red"]
        }
    }

print(fruit_dict)
```

##### dict 的合併

```python
# dict 的合併
# dict.update()
fruit_dict2 = {
    "kiwi":{
            "price": 30,
            "amount": 20,
            "color": ["green"]
        } 
}

fruit_dict.update(fruit_dict2)
print(fruit_dict)
```

```python
print(fruit_dict)
fruit_dict3 = {
    # 拆解dict放進來
    **fruit_dict,
     "kiwi":{
            "price": 100,
            "amount": 50,
            "color": ["yellow"]
        } 
}

print(fruit_dict3)
```

##### 另外的 print 方式 `pprint` (pretty-print) 

```python
# 從 Module模組 引入 function套件
# pprint打印出來的資料 沒有按照順序
from pprint import pprint
pprint(fruit_dict3)
```

```python
'''
新增顏色
apple 有新的品種 顏色要有黃色
tomato 有綠色
'''

# color:值是list 所以可以用append增加
# 重複執行數值會再重複
fruit_dict3["apple"]["color"].append("yellow")
fruit_dict3["tomato"]["color"].append("green")

pprint(fruit_dict3)
```
