##### 將 data.txt 的資料分類成另外三個檔案，分別為 cc.txt 、 css.txt 、 c.txt  

```python
def insert_data(file_path, info):
    with open(file_path, 'a') as f:
        f.write(info)

with open('data.txt', 'r') as f:
    data_list = f.readlines()

for data in data_list:
    data_without_n = data.strip('\n')
    if data_without_n.split('.')[-1] == 'css':
        insert_data('css.txt', data)
    elif data_without_n.split('.')[-1] == 'cc':
        insert_data('cc.txt', data)
    elif data_without_n.split('.')[-1] == 'c':
        insert_data('c.txt', data)
```
